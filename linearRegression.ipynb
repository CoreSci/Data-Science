```python
#Here's a simple example of linear regression in Python using the popular library, scikit-learn
#In this example, we generate random data points, fit a linear regression model, and then plot the data along with the regression line.
#The LinearRegression class from scikit-learn is used for training and making predictions.
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Generate some random data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
# Make predictions
X_new = np.array([[0], [2]])
y_pred = model.predict(X_new)
```

```python
# Plot the data and the regression line
plt.scatter(X, y)
plt.plot(X_new, y_pred, "r-")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
